```python
%matplotlib inline

from image_generator import generate_image_per_style, generate_image_per_loss, STYLE_EMBEDDINGS
from utils import load_embedding_bin

import os
from pathlib import Path
import torch
```

```python
prompt = "Monkeys fighting"
additional_prompt = "duing a snow time"
```

```python


for style_name in STYLE_EMBEDDINGS:
    print(style_name)
    style_file = STYLE_EMBEDDINGS[style_name]
    file_path = os.path.join(os.getcwd(), style_file)
    embedding = load_embedding_bin(file_path)
    
    style_key = Path(style_file).stem
    style_key = style_key.replace("_", "-")
    fig = generate_image_per_style(
        prompt,
        embedding,
        23,
        f"<{style_key}>"
    )
    fig.show()
        
    generate_image_per_loss(
        prompt, embedding, 3, f"<{style_key}>", "cosine_loss", additional_prompt)

```
